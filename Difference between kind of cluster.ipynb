```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

#DBSCAN clustering
from sklearn.cluster import DBSCAN

#Agglomerative Clustering 
from sklearn.cluster import AgglomerativeClustering 

#kmeans clusteing
from sklearn.cluster import KMeans
```

```python
df1 = pd.read_csv('data/data (1).csv')
df2 = pd.read_csv('data/data (2).csv')
df3 = pd.read_csv('data/data (3).csv')
df4 = pd.read_csv('data/data (4).csv')
```

## Data 1 :

```python
df1.head()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df1,x='x',y='y')
plt.show()
```

#### DBSCAN :

```python
db1 = DBSCAN(eps=15,min_samples=6,metric='euclidean').fit(df1)
df1['label']= db1.labels_
db1.labels_
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df1,x='x',y='y',hue='label', palette="Set2")
plt.show()
```

#### Agglomerative:

```python
agglom = AgglomerativeClustering(n_clusters =2, linkage = 'average')
agglom.fit(df1)
```

```python
df1['label1'] = agglom.labels_
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df1,x='x',y='y',hue='label1',palette="Set2")
plt.show()
```

#### K means :

```python
kmeans=KMeans(n_clusters=2, random_state=0).fit(df1)
df1['label3'] = kmeans.labels_
k_means_cluster_centers = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data =df1,x= 'x', y= 'y', hue= 'label3' )
plt.scatter(k_means_cluster_centers[:,0],k_means_cluster_centers[:,1], color = 'r')
plt.show()
```
